```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import chardet
```

```python
file_path1 = '/content/coral.csv'
file_path2 = '/content/DatasetEcoral.csv'

# Primeiro Dataset
dataset1 = pd.read_csv(file_path1, on_bad_lines='skip', low_memory=False)

# Segundo Dataset
with open(file_path2, 'rb') as f:
    encoding = chardet.detect(f.read())['encoding']
dataset2 = pd.read_csv(file_path2, delimiter=';', encoding=encoding, low_memory=False)
```

<h1 style="text-align: center;">Branqueamento de Corais no mundo: Motivo</h1>
<p style="text-align: justify;">
A morte dos corais relacionada ao aquecimento dos oceanos ocorre por um processo chamado de branqueamento, causado quando a temperatura da água fica 1°C acima da média a longo prazo. Diante desse estresse térmico, os corais expulsam as algas que vivem em seus tecidos, as zooxantelas, o que leva à perda de suas cores vibrantes, revelando seu esqueleto calcário, que é branco. Por meio da fotossíntese, as zooxantelas garantem nutrientes aos corais. Sem essas algas, eles conseguem sobreviver por algum tempo, mas ficam mais sujeitos a doenças e à morte.
</p>
<p style="text-align: justify;">
Nos últimos 12 meses a temperatura média da atmosfera terrestre esteve quase 1,5 °C acima dos níveis pré-industriais. Em fevereiro de 2024 a temperatura da superfície dos oceanos, em todo o mundo, foi a maior já registrada. A média para o mês chegou a 21,06°C, quebrando o recorde de agosto de 2023, de 20,98°C, segundo dados do observatório climático europeu Copernicus.
</p>

```python
# Lista única de ocean_name
ocean_names = dataset1['Ocean_Name'].unique()

plt.figure(figsize=(12, 6))
for Ocean_Name in ocean_names:
    # Filtrar dados por ocean_name
    X_subset = dataset1.loc[dataset1['Ocean_Name'] == Ocean_Name, ['Temperature_Kelvin']]
    y_subset = dataset1.loc[dataset1['Ocean_Name'] == Ocean_Name, 'Percent_Bleaching']

    # Remover linhas com valores não numéricos
    X_subset = X_subset.apply(pd.to_numeric, errors='coerce')
    y_subset = y_subset.apply(pd.to_numeric, errors='coerce')

    # Remover linhas com valores NaN resultantes da conversão
    X_subset = X_subset.dropna()
    y_subset = y_subset.dropna()

    # Garantir que ambos X e y têm o mesmo índice após a remoção de NaN
    common_index = X_subset.index.intersection(y_subset.index)
    X_subset = X_subset.loc[common_index]
    y_subset = y_subset.loc[common_index]

    # Converter temperatura de Kelvin para Celsius
    X_subset['Temperature_Celsius'] = X_subset['Temperature_Kelvin'] - 273.15

    plt.scatter(X_subset['Temperature_Celsius'], y_subset, label=Ocean_Name, alpha=0.7)

plt.xlabel('Temperatura (Celsius)')
plt.ylabel('Porcentagem de Branqueamento')
plt.title('Relação entre a Temperatura da Água (°C) e o Branqueamento de Corais, por Localização Oceânica')
plt.legend()
plt.show()
```

```python
import warnings

# Filtrar os dados para os anos de interesse
filtered_data = dataset2[dataset2['Ano_Estimado_de_crescimento'].isin([2025, 2035, 2075])]

# Desativar os avisos de cópia de DataFrame
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)

filtered_data['Estimativa_Km2_Valor'] = filtered_data['Estimativa_Km2_Valor'].astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
filtered_data['Estimativa_Km2_Valor'] = filtered_data['Estimativa_Km2_Valor']

# Convertendo 'Ano_Estimado_de_crescimento' para inteiros para garantir que o eixo x mostre apenas os anos
filtered_data['Ano_Estimado_de_crescimento'] = filtered_data['Ano_Estimado_de_crescimento'].astype(int)

grouped_data = filtered_data.groupby(['Ano_Estimado_de_crescimento', 'Especies_de_Corais'])['Estimativa_Km2_Valor'].sum().unstack()

ax = grouped_data.plot(kind='bar', figsize=(18, 6), width=0.9)

plt.xlabel('Ano')
plt.ylabel('Valor Estimado (em milhões)')
plt.title('Estimativa de Retorno Econômico para o Turismo')
plt.xticks(rotation=0)
plt.legend(title='Espécies de Corais', bbox_to_anchor=(1.05, 1), loc='upper left')

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

plt.tight_layout()
plt.show()
```

<h1 style="text-align: center;">Crescimento de Fragmentos de Corais</h1>
<p style="text-align: justify;">
O crescimento dos corais pode variar bastante dependendo da espécie, das condições ambientais e da localização.
Esses números são médias e podem ser influenciados por vários fatores, incluindo:

*   Temperatura da Água: Corais crescem melhor em águas quentes, mas temperaturas muito altas podem causar estresse e branqueamento.

*   Luz Solar: Corais dependem da luz solar para a fotossíntese realizada pelas algas zooxantelas.

*   Qualidade da Água: Níveis de nutrientes, pH e salinidade afetam o crescimento dos corais.

*   Disponibilidade de Espaço: Corais precisam de espaço para crescer, e a competição com outras espécies pode limitar seu crescimento.
</p>
<p style="text-align: justify;">
O monitoramento regular e a manutenção de um ambiente saudável são cruciais para promover o crescimento adequado dos corais, especialmente em aquários ou em programas de restauração de recifes.
</p>

<h2>Importância dos Corais para o Meio Ambiente</h2>

<h3>Pequenos Corais (Centímetros)</h3>
<ol>
  <li><strong>Habitat para Microorganismos:</strong> Mesmo pequenos corais fornecem habitat para microrganismos marinhos, pequenos crustáceos e outros invertebrados.</li>
  <li><strong>Filtragem da Água:</strong> Corais pequenos ajudam a filtrar a água, removendo partículas e contribuindo para a qualidade da água.</li>
</ol>

<h3>Corais Médios (Dezenas de Centímetros)</h3>
<ol>
  <li><strong>Estrutura do Recife:</strong> À medida que crescem, os corais começam a formar estruturas mais complexas, proporcionando abrigo e áreas de reprodução para uma variedade maior de espécies marinhas.</li>
  <li><strong>Biodiversidade:</strong> Corais médios aumentam a biodiversidade local ao fornecer habitats para peixes e outros organismos marinhos.</li>
  <li><strong>Proteção Costeira:</strong> Contribuem para a formação de recifes que ajudam a dissipar a energia das ondas, protegendo as costas da erosão.</li>
</ol>

<h3>Grandes Corais (Metros)</h3>
<ol>
  <li><strong>Ecossistemas Complexos:</strong> Corais grandes formam a base de ecossistemas de recifes de coral complexos, que são alguns dos habitats mais biodiversos do planeta.</li>
  <li><strong>Ciclagem de Nutrientes:</strong> Eles desempenham um papel crucial na ciclagem de nutrientes no ambiente marinho.</li>
  <li><strong>Serviços Ecossistêmicos:</strong> Proporcionam serviços ecossistêmicos significativos, como sustento para a pesca, turismo e proteção costeira em larga escala.</li>
  <li><strong>Armazenamento de Carbono:</strong> Através da construção de suas estruturas de carbonato de cálcio, os corais ajudam a sequestrar carbono.</li>
</ol>

<h3>Importância Independente do Tamanho</h3>
<ul>
  <li><strong>Indicação de Saúde do Ecossistema:</strong> A presença de corais, independentemente do tamanho, é um indicador de um ecossistema marinho saudável.</li>
  <li><strong>Interações Simbióticas:</strong> Mesmo corais pequenos têm relações simbióticas com algas zooxantelas, que são fundamentais para a produção de oxigênio e remoção de dióxido de carbono.</li>
</ul>
<br/>
<p>Portanto, qualquer coral, desde os estágios iniciais de crescimento, já está contribuindo para o meio ambiente. No entanto, à medida que crescem e se tornam parte de estruturas maiores e mais complexas, sua contribuição aumenta significativamente em termos de biodiversidade, proteção costeira e serviços ecossistêmicos.</p>
